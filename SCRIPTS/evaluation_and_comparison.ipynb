```python
#load results
image_only = pd.read_csv("../outputs/image_only_metrics.csv")
multimodal = pd.read_csv("../outputs/multimodal_metrics.csv")
```

```python
#compare
comparison = pd.DataFrame({
    "Model": ["Image Only", "Image + Metadata"],
    "Accuracy": [image_only["accuracy"][0], multimodal["accuracy"][0]],
    "F1": [image_only["f1"][0], multimodal["f1"][0]]
})

comparison
```

```python
#visualization
comparison.set_index("Model").plot(kind="bar")
plt.title("Model Performance Comparison")
plt.show()
```
